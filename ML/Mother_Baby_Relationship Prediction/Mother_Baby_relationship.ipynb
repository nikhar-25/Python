```python
df2['Crying seconds'] = df2['Crying seconds'].fillna(df2['Crying seconds'].mean())
df2['Bottle amount'] = df2['Bottle amount'].fillna(df2['Bottle amount'].mean())
df2['Pumping Left'] = df2['Pumping Left'].fillna(df2['Pumping Left'].mean())
df2['PumpingRight'] = df2['PumpingRight'].fillna(df2['PumpingRight'].mean())
df2['SleepingSecond'] = df2['SleepingSecond'].fillna(df2['SleepingSecond'].mean())
```

```python
df2
```

```python
corr = df2.corr()
plt.figure(figsize=(8,8))
ax=sns.heatmap(corr,annot=True,linewidth=.5,fmt='0.3f',
               vmin=-1,vmax=1,cmap="coolwarm")
```
